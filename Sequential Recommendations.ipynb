```python
from sklearn.metrics.pairwise import cosine_similarity

def item_similarity(item_id_1, item_id_2):
    item1_vector = user_item_matrix[item_id_1].values.reshape(1, -1)
    item2_vector = user_item_matrix[item_id_2].values.reshape(1, -1)
    
    return cosine_similarity(item1_vector, item2_vector)[0, 0]
```

```python
def predict_rating(user_id, movie_id):
    rated_movies = user_item_matrix.loc[user_id]
    
    numerator = 0
    denominator = 0

    for i, rating in rated_movies.iteritems():
        if rating != 0 and i != movie_id:
            similarity = item_similarity(i, movie_id)
            numerator += similarity * rating
            denominator += abs(similarity)

    if denominator == 0:
        return 0
    else:
        return numerator / denominator
```

```python
user_ids_to_predict = [10, 70, 90]

# Dictionary to store cumulative predicted ratings for each movie
cumulative_ratings = {}

# Loop through each user ID
for user_id in user_ids_to_predict:
    # Get movies that the user hasn't rated
    unrated_movies = user_item_matrix.loc[user_id][user_item_matrix.loc[user_id] == 0].index

    # Predict ratings for unrated movies
    predictions = {movie_id: predict_rating(user_id, movie_id) for movie_id in unrated_movies}

    # Accumulate predicted ratings
    for movie_id, predicted_rating in predictions.items():
        if movie_id not in cumulative_ratings:
            cumulative_ratings[movie_id] = predicted_rating
        else:
            cumulative_ratings[movie_id] += predicted_rating
```

```python
def calculate_group_average_score(user_movie_ratings, user_ids):
    # Calculate average ratings for each movie across the specified user group
    group_average_ratings = {}

    for user_id in user_ids:
        for movie_id, predicted_rating in user_movie_ratings[user_id]['ratings'].items():
            if movie_id not in group_average_ratings:
                group_average_ratings[movie_id] = [predicted_rating]
            else:
                group_average_ratings[movie_id].append(predicted_rating)

    # Calculate the group average score for each movie
    for movie_id, ratings_list in group_average_ratings.items():
        group_average_ratings[movie_id] = sum(ratings_list) / len(ratings_list)

    return group_average_ratings
```

```python
def calculate_group_least_misery_score(user_movie_ratings, user_ids):
    # Calculate the minimum rating for each movie across the specified user group
    group_least_misery_ratings = {}

    for user_id in user_ids:
        for movie_id, predicted_rating in user_movie_ratings[user_id]['ratings'].items():
            if movie_id not in group_least_misery_ratings:
                group_least_misery_ratings[movie_id] = [predicted_rating]
            else:
                group_least_misery_ratings[movie_id].append(predicted_rating)

    # Calculate the group least misery score for each movie
    for movie_id, ratings_list in group_least_misery_ratings.items():
        group_least_misery_ratings[movie_id] = min(ratings_list)

    return group_least_misery_ratings
```

```python
def predict_score_sq(active_user, item, user_item_matrix_n, correlation_matrix):
    
    active_user_mean = user_mean_ratings.get(active_user, 0)
    item_ratings = user_item_matrix_n.loc[item].dropna()
    
    user_correlations = correlation_matrix.get(active_user, {})
    
    user_correlations=user_correlations.sort_values(ascending=False)
    
    user_correlations=user_correlations[0:51,]
    
    weighted_sum = sum(user_correlations.get(other_user, 0) * (item_rating - user_mean_ratings.get(other_user, 0))
                       for other_user, item_rating in item_ratings.items()
                       if other_user != active_user)
    
    sum_of_correlations = sum(abs(user_correlations.get(other_user, 0))
                              for other_user in item_ratings.index
                              if other_user != active_user)
    
    if sum_of_correlations == 0:
        return active_user_mean
    else:
        prediction = active_user_mean + (weighted_sum / sum_of_correlations)
        
        return  prediction
```

```python
def recommendation_for_sq(user_id, user_item_matrix_n, correlation_matrix):
    user_mean_ratings = user_item_matrix_n.mean(axis=1)
    unrated_movies = user_item_matrix_n.loc[user_item_matrix_n[user_id].isna(), :]
    predictions = {}
    #recommended_movies = {}
    
    for movie_id in unrated_movies.index:
        predicted_rating = predict_score_sq(user_id, movie_id, user_item_matrix_n, correlation_matrix)
        predictions[movie_id] = predicted_rating
    
    recommended_movies = sorted(predictions.items(), key=lambda x: x[1], reverse=True)
    return recommended_movies
```

# Proposed Model

### We have used the round robin method with hybrid sequential model to enhence the efficiency of the hybrid sequential model. This model is used in recommender system. It ensures that each user in the group has an equal opportunity to contribute to the recommendations. 

```python
import itertools
```

### Implementing round robin method

```python
def round_robin_rotation(user_group):
    return itertools.cycle(user_group)
```

```python
def calculate_satisfaction(recommendations, user_ratings):
    return sum(recommendations[movie_id] for movie_id in user_ratings.keys() if movie_id in recommendations)
```

### Aggregating Hybrid Sequential model with round robin method

```python
def hybrid_sequential_round_robin(user_item_matrix, user_movie_ratings, top_n=10, a_values=None):
    # Calculate group average and least misery scores
    group_average_scores = calculate_group_average_score(user_movie_ratings, user_movie_ratings.keys())
    group_least_misery_scores = calculate_group_least_misery_score(user_movie_ratings, user_movie_ratings.keys())

    # Create a dictionary to store hybrid scores
    hybrid_scores = {}

    # Perform round-robin rotation
    user_rotation = round_robin_rotation(user_movie_ratings.keys())

    for _ in range(len(user_movie_ratings)):
        user_id = next(user_rotation)
        user_data = user_movie_ratings[user_id]['ratings']

        for movie_id, predicted_rating in user_data.items():
            average_method_score = group_average_scores[movie_id]
            least_misery_score = group_least_misery_scores[movie_id]

            # Calculate hybrid score using the given equation
            hybrid_score = (1 - a_values[user_id]) * average_method_score + a_values[user_id] * least_misery_score

            if movie_id not in seq_group_rec_key:
                hybrid_scores[movie_id] = hybrid_score
    # Sort movies by hybrid score in descending order
    sorted_hybrid_movies = sorted(hybrid_scores.items(), key=lambda x: x[1], reverse=True)

    # Get the top N movies based on hybrid score
    top_n_hybrid_movies = dict(sorted_hybrid_movies[:top_n])

    return top_n_hybrid_movies
```
