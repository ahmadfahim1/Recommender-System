```python
import pandas as pd
```

```python
data = pd.read_csv("u.data.data", sep="\t", names=['user_id', 'movie_id', 'rating', 'timestamp'])
```

```python
user_item_matrix = data.pivot_table(index='user_id', columns='movie_id', values='rating')
```

```python
user_item_matrix
```

```python
user_item_matrix = user_item_matrix.fillna(0)
```

```python
user_item_matrix
```

```python
user_s = user_item_matrix.corr(method='pearson')
```

```python
user_similarity = pd.DataFrame(user_s, index=user_item_matrix.index, columns=user_item_matrix.index)
```

```python
print("User Similarity Between Users: ")
user_similarity
```

```python
from sklearn.metrics.pairwise import cosine_similarity

def item_similarity(item_id_1, item_id_2):
    item1_vector = user_item_matrix[item_id_1].values.reshape(1, -1)
    item2_vector = user_item_matrix[item_id_2].values.reshape(1, -1)
    
    return cosine_similarity(item1_vector, item2_vector)[0, 0]
```

```python
def predict_rating(user_id, movie_id):
    rated_movies = user_item_matrix.loc[user_id]
    
    numerator = 0
    denominator = 0

    for i, rating in rated_movies.iteritems():
        if rating != 0 and i != movie_id:
            similarity = item_similarity(i, movie_id)
            numerator += similarity * rating
            denominator += abs(similarity)

    if denominator == 0:
        return 0
    else:
        return numerator / denominator
```

```python
user_ids_to_predict = [10, 20, 30]

# Dictionary to store cumulative predicted ratings for each movie
cumulative_ratings = {}

# Loop through each user ID
for user_id in user_ids_to_predict:
    # Get movies that the user hasn't rated
    unrated_movies = user_item_matrix.loc[user_id][user_item_matrix.loc[user_id] == 0].index

    # Predict ratings for unrated movies
    predictions = {movie_id: predict_rating(user_id, movie_id) for movie_id in unrated_movies}

    # Accumulate predicted ratings
    for movie_id, predicted_rating in predictions.items():
        if movie_id not in cumulative_ratings:
            cumulative_ratings[movie_id] = predicted_rating
        else:
            cumulative_ratings[movie_id] += predicted_rating
```

```python
average_ratings = {movie_id: rating / len(user_ids_to_predict) for movie_id, rating in cumulative_ratings.items()}

sorted_average_ratings = sorted(average_ratings.items(), key=lambda x: x[1], reverse=True)
```

```python
top_recommendations = sorted_average_ratings[:10]
print("\nTop 10 Recommended Movies based on Average Ratings:\n")
print("   movie_id,   ratings")
top_recommendations
```

```python
user_movie_ratings = {user_id: {} for user_id in user_ids_to_predict}

# Loop through each user ID
for user_id in user_ids_to_predict:
    # Get movies that the user hasn't rated
    unrated_movies = user_item_matrix.loc[user_id][user_item_matrix.loc[user_id] == 0].index

    # Predict ratings for unrated movies
    predictions = {movie_id: predict_rating(user_id, movie_id) for movie_id in unrated_movies}

    # Store predicted ratings for the current user
    user_movie_ratings[user_id] = predictions
```

```python
user_movie_ratings
```

```python
least_misery_ratings = {}

# Loop through each movie
for movie_id in user_item_matrix.columns:
    # Extract ratings for the current movie from each user
    ratings_for_movie = [user_movie_ratings[user_id].get(movie_id, 0) for user_id in user_ids_to_predict]
    
    # Calculate least misery predicted rating (minimum rating across users)
    least_misery_ratings[movie_id] = min(ratings_for_movie)

# Sorting movies by least misery predicted ratings in descending order
sorted_least_misery_ratings = sorted(least_misery_ratings.items(), key=lambda x: x[1], reverse=True)
```

```python
top_recommendation_least_misery = sorted_least_misery_ratings[:10]
```

```python
print("  movie_id,   Ratings")
top_recommendation_least_misery
```
