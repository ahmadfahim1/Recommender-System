```python
user_item_matrix = data.pivot_table(index='user id', columns='item id', values='rating')
```

```python
user_item_matrix
```

```python
user_item_matrix = user_item_matrix.fillna(0)
```

```python
user_item_matrix
```

```python
user_s = user_item_matrix.corr(method='pearson')
```

```python
user_similarity = pd.DataFrame(user_s, index=user_item_matrix.index, columns=user_item_matrix.index)
```

```python
print("User Similarity Between Users: ")
user_similarity
```

```python
def predict_movie_score(a, p, user_item_matrix, user_similarity, top_n=10):
    # ra = rating of user a
    ra = user_item_matrix.loc[a]

    # 𝒑𝒓𝒆𝒅(𝒂,𝒑)=¯𝒓𝒂+ ∑𝒃 ∈𝑵 𝒔𝒊𝒎(𝒂,𝒃)∗(𝒓𝒃,𝒑−¯𝒓𝒃) /∑𝒃 ∈𝑵 𝒔𝒊𝒎(𝒂,𝒃)
    # x = ∑𝒃 ∈𝑵 𝒔𝒊𝒎(𝒂,𝒃)∗(𝒓𝒃,𝒑−¯𝒓𝒃)
    # y = ∑𝒃 ∈𝑵 𝒔𝒊𝒎(𝒂,𝒃)
    x = 0
    y = 0


    top_similar_users = user_similarity.loc[a].nlargest(top_n).index

    for b in top_similar_users:
        if b == a:
            continue

        rb = user_item_matrix.loc[b]

        # similarity between users a and b
        similarity_ab = user_similarity.loc[a, b]

        prediction_bp = rb[p]

        x += similarity_ab * (prediction_bp - rb.mean())
        y += similarity_ab

    if y == 0:
        return ra.mean()
    else:
        return ra.mean() + x / y
```

```python
user_id = 105
movie_id = 182
predicted_rating = predict_movie_score(user_id, movie_id, user_item_matrix, user_similarity)
print(f'Ranting Prediction for movie-{movie_id} given by user-{user_id}: {predicted_rating}')
```

```python
def similar_user(a, user_similarity, n):
    similar_users = user_similarity.loc[a].drop(a).nlargest(n).index
    return similar_users
    
```

```python
user_id = 1
n = 10
similar_user = similar_user(user_id, user_similarity, n)
print(f'{n} similar user of user-{user_id} are: {similar_user}')
```

```python
# Recommending 10 movies to a user 
# This was done by finding the most similar user based on user similarity then recommending the 10 movies that user given best rating.
def find_most_similar_user(user_id, user_similarity):
    #This function finds the most similar user to the given user id based on user similarity except himself
    most_similar_user = user_similarity.loc[user_id].drop(user_id).idxmax()
    return most_similar_user

def get_top_rated_movies_by_user(user_id, user_item_matrix, most_similar_user, n):
    # getting the most rated movies of the similar user
    similar_user_ratings = user_item_matrix.loc[most_similar_user]
    
    #excluding the movies that the given user already rated
    unrated_movies = similar_user_ratings[user_item_matrix.loc[user_id] == 0]

    # sorting movies by their rating
    movie_suggestion = unrated_movies.sort_values(ascending = False).head(n)

    return movie_suggestion
```

```python
user_id = 10
n = 10
similar_user = find_most_similar_user(user_id, user_similarity)
recommended_movies = get_top_rated_movies_by_user(user_id, user_item_matrix, similar_user, n)
print(f'Recommended Movies for user {user_id} are : {recommended_movies.index}')
```

```python
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np
```

```python
user_item_matrix = data.pivot_table(index='user id', columns='item id', values='rating')
```

```python
user_item_matrix = user_item_matrix.fillna(0)
user_item_matrix
```

```python
user_item_array = np.array(user_item_matrix)
user_item_array
```

```python
def cosine_similarity(matrix):
    num_users = matrix.shape[0]
    similarity_matrix = np.zeros((num_users, num_users))

    for i in range(num_users):
        for j in range(num_users):
                user_i = matrix[i]
                user_j = matrix[j]
                cosine = np.sum(user_i * user_j) / (np.linalg.norm(user_i) * np.linalg.norm(user_j))
                similarity_matrix[i, j] = cosine

    return similarity_matrix
```

```python
user_similarity_m = cosine_similarity(user_item_array)
user_similarity_m
```
